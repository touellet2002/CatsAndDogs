<h1>Imports</h1>

```python
import os
import random

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D

import numpy as np
import matplotlib.pyplot as plt

import cv2
```

<h1>The Dataset</h1>
<p>I took my dataset from Kaggle.  The datset is composed of a <b>training set</b> and a <b>testing set</b>.  Each set are divided in two folders.  One folder contains the images of the cats, and the other one contains the images of the dogs.</p>

```python
TESTING_DATADIR = r"C:\Users\Utilisateur\Desktop\Notebooks\datasets\training_set\training_set"
CATEGORIES = ["cats", "dogs"]

for category in CATEGORIES:
    path = os.path.join(TESTING_DATADIR, category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array, cmap="gray")
        plt.show()
        break
    break
```

```python
print(img_array)
```

```python
IMG_SIZE = 50

resized_img_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
plt.imshow(resized_img_array, cmap="gray")
plt.show()
```

```python
training_data = []

for category in CATEGORIES:
    path = os.path.join(TESTING_DATADIR, category)
    class_num = CATEGORIES.index(category)
    for img in os.listdir(path):
        try:
            img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
            resized_img_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
            training_data.append([resized_img_array, class_num])
        except Exception as e:
            pass
```

```python
print(len(training_data))
```

```python
#random.shuffle(training_data)
for sample in training_data[:10]:
    print(sample[1])
```

```python
X = []
y = []

for features, label in training_data: 
    X.append(features)
    y.append(label)
```

```python
X = np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
y = np.array(y)
```

```python
gpu_options = tf.compat.v1.GPUOptions(per_process_gpu_memory_fraction=0.5)
sess = tf.compat.v1.Session(config=tf.compat.v1.ConfigProto(gpu_options=gpu_options))
```

```python
X = X/255.0

model = Sequential()
model.add( Conv2D(64, (3, 3), input_shape = X.shape[1:]) )
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)) )
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))

model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss="binary_crossentropy",
             optimizer="adam",
             metrics=['accuracy'])
          
model.fit(X, y, batch_size=32, epochs=12, validation_split=0.2, shuffle=True)
```

```python
model.save("cats-dogs-model.model")
```
